```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.metrics import mean_absolute_percentage_error, make_scorer
from sklearn.model_selection import cross_val_score
import plotly.express as px
from sklearn.model_selection import KFold
```

```python
df=pd.read_csv('final_data.csv')
df
```

```python
#Hyperparameter tuning:  "Number of Estimators"
df_total_estimators=pd.DataFrame(columns=['n','MAPE'])
x=df.iloc[:,:27]
y_total=df.Total
n_array=[]
mean_MAPE_estimator=[]
for i in range(10,210,10):
  k=5
  kf = KFold(n_splits=k, random_state=None)
  MAPE_estimator=[]
  for train_index , test_index in kf.split(x): 
    x_total_train , x_total_test = x.iloc[train_index,:],x.iloc[test_index,:]
    y_total_train , y_total_test = y_total[train_index] , y_total[test_index]
    model_total=RandomForestRegressor(n_estimators=i)
    model_total.fit(x_total_train,y_total_train)
    prediction_total=model_total.predict(x_total_test)
    mape_metric_estimator = mean_absolute_percentage_error(prediction_total,y_total_test)
    MAPE_estimator.append(mape_metric_estimator)
  n_array.append(i)
  mean_MAPE_estimator.append( round( (100*(np.mean(MAPE_estimator))),2)  )
df_total_estimators['n']=n_array
df_total_estimators['MAPE']=mean_MAPE_estimator

fig_estimators= px.line(df_total_estimators,
                     x="n", y="MAPE", title="Hyperparameter Tuning : Number of Estimators",labels={"n":"Number of Estimators","MAPE":"Mean Absolute Percentage error"},width=1000, height=400,
                     template="plotly_white")
fig_estimators.update_layout(yaxis_range=[30,40])
fig_estimators.show()
```

```python
#Hyperparameter tuning: max_depth
df_total_max_depth=pd.DataFrame(columns=['n','MAPE'])
x=df.iloc[:,:27]
y_total=df.Total
max_depth_array=[]
mean_MAPE_max_depth=[]
for i in range(5,15,1):
  k=5
  kf = KFold(n_splits=k, random_state=None)
  MAPE_max_depth=[]
  for train_index , test_index in kf.split(x): 
    x_total_train , x_total_test = x.iloc[train_index,:],x.iloc[test_index,:]
    y_total_train , y_total_test = y_total[train_index] , y_total[test_index]
    model_total=RandomForestRegressor(max_depth=i)
    model_total.fit(x_total_train,y_total_train)
    prediction_total=model_total.predict(x_total_test)
    mape_metric_max_depth = mean_absolute_percentage_error(prediction_total,y_total_test)
    MAPE_max_depth.append(mape_metric_max_depth)
  max_depth_array.append(i)
  mean_MAPE_max_depth.append( round( (100*(np.mean(MAPE_max_depth))),2)  )
df_total_max_depth['n']=max_depth_array
df_total_max_depth['MAPE']=mean_MAPE_max_depth

fig_max_depth= px.line(df_total_max_depth,
                     x="n", y="MAPE", title="Hyperparameter Tuning : Maximum Tree Depth",labels={"n":"Maximum Tree depth","MAPE":"Mean Absolute Percentage error"},width=1000, height=400,
                     template="plotly_white")
fig_max_depth.update_layout(yaxis_range=[30,40])
fig_max_depth.show()
```

```python
#Hyperparameter tuning: min_samples_split
df_total_min_samples_split=pd.DataFrame(columns=['n','MAPE'])
x=df.iloc[:,:27]
y_total=df.Total
min_samples_split_array=[]
mean_MAPE_min_samples_split=[]
for i in range(2,11,1):
  k=5
  kf = KFold(n_splits=k, random_state=None)
  MAPE_min_samples_split=[]
  for train_index , test_index in kf.split(x): 
    x_total_train , x_total_test = x.iloc[train_index,:],x.iloc[test_index,:]
    y_total_train , y_total_test = y_total[train_index] , y_total[test_index]
    model_total=RandomForestRegressor(min_samples_split=i)
    model_total.fit(x_total_train,y_total_train)
    prediction_total=model_total.predict(x_total_test)
    mape_metric_min_samples_split = mean_absolute_percentage_error(prediction_total,y_total_test)
    MAPE_min_samples_split.append(mape_metric_min_samples_split)
  min_samples_split_array.append(i)
  mean_MAPE_min_samples_split.append( round( (100*(np.mean(MAPE_min_samples_split))),2)  )
df_total_min_samples_split['n']=min_samples_split_array
df_total_min_samples_split['MAPE']=mean_MAPE_min_samples_split

fig_min_samples_split= px.line(df_total_min_samples_split,
                     x="n", y="MAPE", title="Hyperparameter Tuning : Minimum number of  samples for splitting an internal node",labels={"n":"Minimum samples","MAPE":"Mean Absolute Percentage error"},width=1000, height=300,
                     template="plotly_white")
fig_min_samples_split.update_layout(yaxis_range=[30,40])
fig_min_samples_split.show()
```

```python
#Hyperparameter tuning: min_samples_leaf
df_total_min_samples_leaf=pd.DataFrame(columns=['n','MAPE'])
x=df.iloc[:,:27]
y_total=df.Total
min_samples_leaf_array=[]
mean_MAPE_min_samples_leaf=[]
for i in range(1,11,1):
  k=5
  kf = KFold(n_splits=k, random_state=None)
  MAPE_min_samples_leaf=[]
  for train_index , test_index in kf.split(x): 
    x_total_train , x_total_test = x.iloc[train_index,:],x.iloc[test_index,:]
    y_total_train , y_total_test = y_total[train_index] , y_total[test_index]
    model_total=RandomForestRegressor(min_samples_leaf=i)
    model_total.fit(x_total_train,y_total_train)
    prediction_total=model_total.predict(x_total_test)
    mape_metric_min_samples_leaf = mean_absolute_percentage_error(prediction_total,y_total_test)
    MAPE_min_samples_leaf.append(mape_metric_min_samples_leaf)
  min_samples_leaf_array.append(i)
  mean_MAPE_min_samples_leaf.append( round( (100*(np.mean(MAPE_min_samples_leaf))),2)  )
df_total_min_samples_leaf['n']=min_samples_leaf_array
df_total_min_samples_leaf['MAPE']=mean_MAPE_min_samples_leaf

fig_min_samples_leaf= px.line(df_total_min_samples_leaf,
                     x="n", y="MAPE", title="Hyperparameter Tuning : Minimum Number of Samples at each leaf node",labels={"n":"Minimum samples","MAPE":"Mean Absolute Percentage error"},width=1000, height=300,
                     template="plotly_white")
fig_min_samples_leaf.update_layout(yaxis_range=[30,40])
fig_min_samples_leaf.show()
```

```python
#Hyperparameter tuning: k value for cross validation
df_total_mape_k=pd.DataFrame(columns=['k','MAPE'])
model_total_tuned=RandomForestRegressor(n_estimators=200,max_depth=11,min_samples_split=8,min_samples_leaf=6)
x=df.iloc[:,:27]
y_total=df.Total
k_array=[]
mean_MAPE_k=[]
for i in range(2,21,2):
  k = i
  kf = KFold(n_splits=k, random_state=None)
  MAPE_k = []
  for train_index , test_index in kf.split(x):
    x_total_train , x_total_test = x.iloc[train_index,:],x.iloc[test_index,:]
    y_total_train , y_total_test = y_total[train_index] , y_total[test_index]
    model_total_tuned.fit(x_total_train,y_total_train)
    prediction_total=model_total_tuned.predict(x_total_test)
    mape_metric_k = mean_absolute_percentage_error(prediction_total,y_total_test)
    MAPE_k.append(mape_metric_k)
  k_array.append(k)
  mean_MAPE_k.append( round( (100*(np.mean(MAPE_k))),2)  )
df_total_mape_k['k']=k_array
df_total_mape_k['MAPE']=mean_MAPE_k

fig_k= px.line(df_total_mape_k,
                     x="k", y="MAPE", title="Hyperparameter Tuning : Number of folds",labels={"k":"Number of folds","MAPE":"Mean Absolute Percentage error"},width=1000, height=300,
                     template="plotly_white")
fig_k.show()
```
